```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
movie_ratings = pd.read_csv('Movie-Ratings.csv')
movie_ratings.head()
```

### List of top 10 movie(descending order) in terms of budget

```python
top_movies = movie_ratings.sort_values('Budget (million $)', ascending=False).head(10)
top_movies
```

```python
top_movies.to_csv('top_movies.csv', index=False)
```

### List of top 10 movie(ascending order) in terms of budget


```python
movies_ascendind=movie_ratings.sort_values('Budget (million $)',ascending=1).head(10)
```

```python
movies_ascendind.to_csv('ascendind.csv',index=False)
```

### Which genre is the best in terms of price budget

```python
movie_ratings.groupby('Genre').mean().drop('Year of release',axis = 1).sort_values('Budget (million $)',ascending=True)
```

```python
movie_ratings.groupby('Genre').max()
```
